```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# NORMAL DISTRIBUTION 

normal = np.random.normal(10,1,10000)
normal_mean = np.mean(normal)
normal_std = np.std(normal)

statement = "The mean is {} and the standard deviation is {}"
full_statement = statement.format(round(normal_mean,2),round(normal_std,2))

plt.hist(normal,bins=25, color='green')
plt.axvline(normal_mean, color = 'red',linestyle = 'solid', linewidth=2)
plt.axvline((normal_mean - normal_std), color = "red", linestyle ='dashed',linewidth=2)
plt.axvline((normal_mean + normal_std), color = "red", linestyle ='dashed',linewidth=2)
plt.title("Normal Distribution")
plt.show()
print(full_statement)
```
