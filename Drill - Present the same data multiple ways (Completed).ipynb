```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("C:\\Users\halln\Desktop\THINKFUL\Datasets\crime_rates\crime.csv")
raman_data = pd.read_csv('C:\\Users\halln\Desktop\THINKFUL\Datasets\Raman\Raman.csv')
dog_data = pd.read_csv('C:\\Users\halln\Desktop\THINKFUL\Datasets\dog_data\dog_data.csv')
df = pd.DataFrame(data)
```

```python
# QUESTION 1 PLOT ONE VARIABLE IN 4 DIFFERENT WAYS

sns.set(style="dark")

fig = plt.figure(figsize=(15,15))
fig.suptitle("RAMAN SPECTRUM")

plt.subplot(221)
g1 = sns.lineplot(y='DM206',
                 x='ramanShift',
                 data=raman_data,
                 color='red')
g1.axvline(2500,color='green',linestyle='--')
g1.axvline(3000,color='green',linestyle='--')
g1.set_xlim(500,3500)
g1.set_ylim(-700,200)
g1.set_xlabel("")
g1.set_ylabel("Intensity")
g1.xaxis.tick_top()
g1.invert_yaxis()

plt.subplot(222)
g2 = sns.lineplot(y='DM206',
                 x='ramanShift',
                 data=raman_data,
                 color='orange')
g2.axvline(2675,color='green',linestyle='--')
g2.axvline(2750,color='green',linestyle='--')
g2.set_xlim(2500,2800)
g2.set_ylim(-700,200)
g2.set_xlabel("")
g2.set_ylabel("")
g2.yaxis.tick_right()
g2.xaxis.tick_top()
g2.invert_yaxis()

plt.subplot(223)
g3 = sns.lineplot(y='DM206',
                 x='ramanShift',
                 data=raman_data,
                 color='gold')
g3.axvline(2705,color='green',linestyle='--')
g3.axvline(2730,color='green',linestyle='--')
g3.set_xlim(2675,2750)
g3.set_ylim(-700,100)
g3.set_xlabel(" Raman Shift ($cm^{-1}$)")
g3.set_ylabel("Intensity")
g3.invert_yaxis()

plt.subplot(224)
g4 = sns.lineplot(y='DM206',
                 x='ramanShift',
                 data=raman_data,
                 color='darkgreen')

g4.set_xlim(2705,2730)
g4.set_ylim(-700,50)
g4.set_xlabel(" Raman Shift ($cm^{-1}$)")
g4.set_ylabel("")
g4.text(2723,-275, "2723 $cm^{-1}$", horizontalalignment='center', size='medium', color='black', weight='bold')
g4.text(2713,-540, "2713 $cm^{-1}$", horizontalalignment='center', size='medium', color='black', weight='bold')
g4.yaxis.tick_right()
g4.invert_yaxis()
```

```python
# QUESTION 2 TAKE 2 CONTINUOUS VARIABLES AND PLOT IN 3 DIFFERENT WAYS

# Subset crime rate data into data for Raleigh and Charlotte 
raleigh = data[ data['agency_jurisdiction'] == 'Raleigh, NC']
charlotte = data[ data['agency_jurisdiction'] == 'Charlotte, NC']
```

```python
# Create Line Plot
fig=plt.figure(figsize=(10,5))
sns.set(style="whitegrid")  

graph=sns.lineplot(x='report_year',
                    y='violent_crimes',
                    data=raleigh,
                    color='orange')
graph=sns.lineplot(x='report_year',
                    y='violent_crimes',
                    data=charlotte,
                    color='green')
graph.set_title("VIOLENT CRIME INCREASE FROM 1975 TO 2015")
graph.legend(("Raleigh","Charlotte"))
graph.set_xlabel("Year")
graph.set_ylabel("Violent Crimes")
plt.show()
```

```python
# Create Linear Regression Plots for Raleigh/Charlotte Violent Crimes Committed per Year
sns.set(style='dark')

lin_reg = sns.lmplot(x='report_year',
           y='violent_crimes',
           data=raleigh)
fig=lin_reg.fig
fig.suptitle("Trend in Violent Crime for Raleigh",fontsize=12)
plt.xlabel("Year")
plt.ylabel("Violent Crimes Committed Per Year")

lin_reg2 = sns.lmplot(x='report_year',
                 y='violent_crimes',
                 data=charlotte)
fig=lin_reg2.fig
fig.suptitle("Trend in Violent Crime for Charlotte",fontsize=12)
plt.xlabel("Year")
plt.ylabel("Violent Crimes Committed Per Year")
plt.show()
```

```python
# Create Barplot 
fig = plt.figure(figsize=(20,5))
sns.set_context("paper")

plt.subplot(121)
chart = sns.barplot(x='report_year',
            y='violent_crimes',
            data=raleigh,
            color='black')
chart.set_title("Violent Crimes Per Year in Raleigh",size='large')
chart.set_xlabel("Year")
chart.set_xticklabels(chart.get_xticklabels(),rotation=90,ha="right")
chart.set_ylabel("Violent Crimes Committed Per Year")

plt.subplot(122)
chart = sns.barplot(x='report_year',
            y='violent_crimes',
            data=charlotte,
            color='black')
chart.set_title("Violent Crimes Per Year in Charlotte",size='large')
chart.set_xlabel("Year")
chart.set_xticklabels(chart.get_xticklabels(),rotation=90,ha="right")
chart.set_ylabel("Violent Crimes Committed Per Year")
plt.show()
```

```python
# QUESTION 3 PART 2 - Plot a categorical variable 6 ways

# Create Barplot

adopted = dog_data[dog_data['outcome_type']=='Adoption']

figure = plt.figure(figsize=(10,10))
figure.suptitle("Length of Stay in Shelter by Animal Type")
sns.set(style='ticks')

plt1 = sns.barplot(x='animal_type',y='time_in_shelter_days',data=adopted,palette='winter')
plt.ylabel("Length of Stay in Shelter (Days)")
plt.xlabel("Animal Type")
plt.xticks(rotation='vertical')
```

```python
# Create Differently oriented barplot

figure = plt.figure(figsize=(10,10))
figure.suptitle("Length of Stay in Shelter by Animal Type")
sns.set(style='darkgrid')

plt2 = sns.barplot(y='animal_type',x='time_in_shelter_days',data=adopted,palette='summer')
plt.ylabel("Animal Type")
plt.xlabel("Length of Stay in Shelter (Days)")
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Create Cat Plot
figure = plt.figure(figsize=(10,10))
sns.set(style='dark')

plt3 = sns.catplot(y='animal_type',x='time_in_shelter_days',data=adopted,palette="twilight")
plt.title("Length of Stay in Shelter by Animal Type")
plt.ylabel("Animal Type")
plt.xlabel("Length of Stay in Shelter (Days)")
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Create Cat Plot
figure = plt.figure(figsize=(10,10))
figure.suptitle("Length of Stay in Shelter by Animal Type")
sns.set(style='whitegrid')

plt4 = sns.catplot(x='animal_type',y='time_in_shelter_days',data=adopted)
plt.title("Length of Stay in Shelter by Animal Type")
plt.ylabel("Animal Type")
plt.xlabel("Length of Stay in Shelter (Days)")
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Create Boxplot
figure = plt.figure(figsize=(10,10))
sns.set(style='darkgrid')

plt5 = sns.boxplot(x='animal_type',y='time_in_shelter_days',data=adopted,palette='autumn')
plt.title("Length of Stay in Shelter by Animal Type")
plt.ylabel("Length of Stay in Shelter (Days)")
plt.xlabel("Animal Type")
plt.ylim(0,400)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Create Boxplot
figure = plt.figure(figsize=(10,10))
sns.set(style='whitegrid')

plt6 = sns.boxplot(y='animal_type',x='time_in_shelter_days',data=adopted,palette='gray')
plt.title("Length of Stay in Shelter by Animal Type")
plt.xlabel("Length of Stay (Days)")
plt.ylabel("Animal Type")
plt.xticks(rotation='vertical')
plt.xlim(0,200)
plt.show()
```

# DISCUSSION

Each of these plots have their pros and cons.  For showing a distribution I personally find boxplots to be the most effective because they seem to convey the most information with the least amount of space and as simply as possible.  They allow someone to easily see how the data clumps, as well as if there are any major outliers.  The cat plots and the barplots are able to convey some of the same information, but for most types of numerical data I would still choose a boxplot.  The cat plots showed how the data clumped, but with the amount of data I was plotting the area towards the bottom was so full that I was unable to distinguish single dots from the rest.
For showing a linear trend over time I found a lineplot to be the most effective, although I believe that using a barplot can be just as effective in most situations, and in some cases can present the data better.
