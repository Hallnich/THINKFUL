```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read in data from kaggle
student_data = pd.read_csv('C:\\Users\halln\Desktop\THINKFUL\Datasets\Students_Performance\StudentsPerformance.csv')
df = pd.DataFrame(student_data)
```

```python
# Sort data into different levels of education

high_school = df[df['parental level of education'] == 'high school']
some_college = df[df['parental level of education'] == 'some college']
associates = df[df['parental level of education'] == 'associate\'s degree']
bachelors = df[df['parental level of education'] == 'bachelor\'s degree']
masters = df[df['parental level of education'] == 'master\'s degree']

# Create boxplots for math scores

scores = (high_school['math score'],some_college['math score'],associates['math score'],bachelors['math score'],masters['math score'])
labels = ("High School","Some College","Associates","Bachelors","Masters")
plt.boxplot(scores,labels=labels)
plt.xlabel("EDUCATION LEVEL")
plt.ylabel("MATH SCORE")
plt.title("MATH SCORES BY EDUCATION LEVEL")
plt.show()

# Create boxplots for writing scores

scores = (high_school['writing score'],some_college['writing score'],associates['writing score'],bachelors['writing score'],masters['writing score'])
labels = ("High School","Some College","Associates","Bachelors","Masters")
plt.boxplot(scores,labels=labels)
plt.xlabel("EDUCATION LEVEL")
plt.ylabel("WRITING SCORE")
plt.title("WRITING SCORES BY EDUCATION LEVEL")
plt.show()

# Create boxplots for reading scores

scores = (high_school['reading score'],some_college['reading score'],associates['reading score'],bachelors['reading score'],masters['reading score'])
labels = ("High School","Some College","Associates","Bachelors","Masters")
plt.boxplot(scores,labels=labels)
plt.xlabel("EDUCATION LEVEL")
plt.ylabel("READING SCORE")
plt.title("READING SCORES BY EDUCATION LEVEL")
plt.show()

names = ('High School','Some College','Associates','Bachelors','Masters')
counts = pd.DataFrame(index=names)
counts['Total Entries'] = (len(high_school),len(some_college),len(associates),len(bachelors),len(masters))
print(counts)


```

The code above created plots which show the spread of the scores in each subject based on the parental level of education.   All together I was very suprised to find that the scores recieved were a much more consistent spread than I expected despite the differences in parental level of education.  Based on the writing scores plot, both the range of scores decreases and the median scores increase as parental education level increases.  The reading scores plot also showed that the median scores increased as parental education level increased.  Intersetingly, the math scores plot showed that while the median scores increased as parental educaiton level increased, the maximum score obtained by a masters student was lower than the maximum score obtained for any other education level.  However, this could be due to the fact that there are substantially less students whose parents obtained a masters when compared to the other education levels

```python
# Sort data by gender 

males = df[df['gender'] == 'male']
females = df[df['gender'] == 'female']

# Create boxplots comparing math scores

scores = (males['math score'],females['math score'])
labels = ("Male","Female")
plt.boxplot(scores,labels=labels)
plt.xlabel("GENDER")
plt.ylabel("MATH SCORE")
plt.title("MATH SCORES")
plt.show()

# Create boxplots comparing writing scores

scores = (males['writing score'],females['writing score'])
labels = ("Male","Female")
plt.boxplot(scores,labels=labels)
plt.xlabel("GENDER")
plt.ylabel("WRITING SCORE")
plt.title("WRITING SCORES")
plt.show()

# Create boxplots comparing reading scores

scores = (males['reading score'],females['reading score'])
labels = ("Male","Female")
plt.boxplot(scores,labels=labels)
plt.xlabel("GENDER")
plt.ylabel("READING SCORE")
plt.title("READING SCORES")
plt.show()

names = ("Male","Female")
count = pd.DataFrame(index=names)
count['Total Entries'] = (len(males),len(females))
print(count)
```

The code above generated box and whisker plots to determine if there were any differences between genders in the scores obtained for each subject matter.  Based on the plots generated, it seems that male scores tend to cluster higher in math, but the female scores tend to cluster higher in the reading and writing scores.
